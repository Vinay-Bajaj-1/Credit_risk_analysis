#Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(main_df, test_size = 0.3, random_state = 1)
print(train.shape, test.shape)
```

```python
#handling outliers

train = train[train['annual_inc'] <= 250000]
train = train[train['dti'] <= 50]
train = train[train['open_acc'] <= 40]
train = train[train['total_acc'] <= 80]
train = train[train['revol_util'] <= 120]
train = train[train['revol_bal'] <= 250000]

print(train.shape)
```

```python
#normalizing data

x_train, y_train = train.drop('loan_status', axis = 1), train.loan_status
x_test, y_test = test.drop('loan_status', axis = 1), test.loan_status
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train = np.array(x_train).astype(np.float32)
x_test = np.array(x_test).astype(np.float32)
y_train = np.array(y_train).astype(np.float32)
y_test = np.array(y_test).astype(np.float32)
```

# ANN model

```python
x_train.shape, y_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, BatchNormalization, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.metrics import AUC
from sklearn.metrics import classification_report
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
i = Input(shape=(x_train.shape[1],))

x = Dense(150, activation = 'relu')(i)
x = BatchNormalization()(x)
x = Dropout(0.1)(x)

x = Dense(150, activation = 'relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.1)(x)

x = Dense(150, activation = 'relu')(x)
x = BatchNormalization()(x)


x = Dense(1, activation = 'sigmoid')(x)

model = Model(i, x)
```

```python
early_stopping = EarlyStopping(monitor='val_AUC', patience=5, restore_best_weights=True)
model.compile(optimizer = Adam(0.001), loss = 'binary_crossentropy', metrics = [AUC(name = 'AUC')])
```

```python
model.summary()
```

```python
r = model.fit(x_train, y_train, epochs = 40, batch_size = 32 ,validation_data = (x_test, y_test), callbacks=[early_stopping])
```

```python
plt.plot(r.history['AUC'], label = 'AUC')
plt.plot(r.history['val_AUC'], label = 'val_AUC')
plt.legend()
```

```python
y_test_pred = model.predict(x_test).round()
```

```python
report = classification_report(y_test, y_test_pred)
print(report)
```

# XGBoost

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 200, 300],
}
```

```python
xgb_classifier = xgb.XGBClassifier(objective='binary:logistic', random_state=42)

grid_search = GridSearchCV(estimator=xgb_classifier, param_grid=param_grid, cv=3, scoring='accuracy')
```

```python
grid_search.fit(x_train, y_train)
```

```python
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_
```

```python
y_pred = best_estimator.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Best Model Accuracy:", accuracy)
```

```python
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```
